```python
# Importing Libraries
import pandas as pd
from datasets import load_dataset
import matplotlib.pyplot as plt

# Loading Data
dataset = load_dataset('lukebarousse/data_jobs')
df = dataset['train'].to_pandas()

# Data Cleanup
df['job_posted_date'] = pd.to_datetime(df['job_posted_date'])
```

```python
# Return a random sample of items from a DataFrame
df.sample()
```

```python
# n= is number of items to return
df.sample(n=5)
```

```python
# Fraction of rows to return (e.g., frac=0.1 returns 10% of the data)
df.sample(frac=0.1)
```

```python
# random_state= is Seed for the random number generator (useful for reproducibility).
df.sample(n=5, random_state=42)
```

```python
# Use copy() method to create completely independent copy of the original DataFrame
df2 = df
```

```python
df2['company_name'] = 'df2 changes affect original df'
df2['company_name']
```

```python
# Notice how df (original dataframe) was modified, even though we only changed df2
df['company_name']
```

```python
# Now, using copy()
df2 = df.copy()
```

```python
df2['company_name'] = 'df2 created by independent copy of original dataframe'
df2['company_name']
```

```python
# This time, df (original dataframe) remains unchanged.
df['company_name']
```
